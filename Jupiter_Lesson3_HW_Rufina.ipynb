- lat - Широтма
- long - Долгота

### 1.1 Создать новый признак price_per_sq_lot, 
который будет содержать стоимость за один кв. метр общей площади

Данной задачи нет в списке домашних задач, но была у вас в презентации во 2 семинаре.

```python
df['price_per_sq_lot'] = df['price'] / df['sqft_lot']
df.head()
```

### 1.2 Создать новый признак delta_renovated, 
который будет содержать разницу в годах между годом реновацией дома и годом постройки дома

```python
df['delta_renovated'] = df['yr_renovated'] - df['yr_built']
df.head()
```

Если реновации дома не было, то в новом признаке поставьте 0

```python
df['delta_renovated'] = df['delta_renovated'].apply(lambda x: x if x > 0 else 0)
df.head()
```

### 1.3 Создайте признаки года продажи, месяца продажи

Создаю графу date_1, где указаны только год, месяц, день.

```python
df['date_1'] = df['date'].apply(lambda x: x.split('T')[0])
df.head()
```

 Преобразовываем в данные в этой графе в datetime.

```python
import datetime
```

```python
df['date_1']=pd.to_datetime(df['date_1'], format='%Y-%m-%d')
df.head()
```

Выводим в отдельные графы признаки года (year_sell) и месяца (month_sell) продажи

```python
df['year_sell']=pd.DatetimeIndex(df['date_1']).year
df.head()
```

```python
df['month_sell']=pd.DatetimeIndex(df['date_1']).month
df.head()
```

### 1.4 Удалите признаки date, zipcode, lat, long

```python
df.drop(columns=['date', 'zipcode', 'lat', 'long'], inplace=True)
df.head()
```
